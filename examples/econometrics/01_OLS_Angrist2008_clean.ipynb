## 理论要点（精简）

考虑线性模型：\
$$Y_i = \beta_0 + \beta_1 X_{1i} + \beta_2 X_{2i} + \epsilon_i$$
OLS 闭式解：$$\hat{\beta} = (X'X)^{-1}X'Y$$
因果解释需要满足条件外生性（Conditional Independence）等假设；参见 Angrist & Pischke (2008)。
