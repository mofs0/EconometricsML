# A02 选题与开题报告

选题决定了你未来两三年的方向，开题报告是你向学术共同体递交的"可行性报告"。  
这一篇解决两个核心问题：**选什么方向**，以及**怎么写开题**。

---


## 一、选题的本质是风险管理

很多同学在选题上有一个误区：越新颖越好、越没人做越独特。

事实恰恰相反。**一个好选题的核心是"可以做出来"**，而不是"听起来很厉害"。

**选题时要评估三件事：**

**1. 数据可不可得？**  
没有数据，任何研究设计都是空中楼阁。在确定方向之前，先确认数据来源。

**2. 方法匹不匹配？**  
你的研究问题需要什么识别策略？你是否有能力掌握这个方法？时间够不够学？

**3. 前人铺没铺路？**  
如果一个方向没有人做过，要么是你发现了真正的空白（极少数），要么是这个问题本身有坑（多数情况）。  
有前人做，意味着有代码可以参考、有数据路径可以追溯、有方法论可以学习。

> 结论：选一个**有基础、有数据、你能做出来**的方向，远比追求新颖但做不完要好。


## 二、如何在课题组内找到好选题

入学初期最有价值的事，不是看教材，而是**把课题组近五年的论文通读一遍**。

这样做你会得到：
- 课题组擅长什么方法、用过什么数据
- 师兄师姐研究了哪些问题，留下了什么"尾巴"没做完
- 哪些问题导师最感兴趣（他会愿意给你更多指导）

**从师兄师姐那里"继承"的正确姿势：**
- 不是直接复制，而是在他们的研究基础上延伸（新时间段、新地区、新调节变量）
- 继承他们处理好的数据（数据清洗是最耗时的工作，能省则省）
- 继承他们的代码框架（回归结构、稳健性检验思路）

这不是抄袭，这是正常的科研积累方式——所有领域的进步都建立在前人工作的基础上。


## 三、开题报告的结构与定位

**一份开题报告的本质**：让评委相信你的研究问题是真实的、你的路径是可行的、你有能力完成它。

**标准结构（五部分）：**

```
第一章  研究背景与意义
第二章  国内外研究现状（文献综述）
第三章  研究设计（内容、数据、方法、技术路线）
第四章  预期贡献与创新点
第五章  参考文献
```

把这五部分写好之后，加上后续的实证结果，它就会自然地成长为你的学位论文。  
所以开题写得越扎实，后面写大论文越省力。


## 四、各部分怎么写

### 4.1 研究背景与意义

这部分要完成一个逻辑跳跃：**从宏大的社会现实，聚焦到你这个具体的研究问题**。

**推荐写法（三层递进）：**

```
第一层：宏观现象或政策背景（2-3 句）
  ——近年来，XX 领域发生了什么重要变化？引用权威数据或政策文件支撑。

第二层：具体问题的重要性（2-3 句）
  ——在这个大背景下，XX 这个具体问题为什么值得关注？有什么经济后果？

第三层：现有研究的不足与本文的切入点（2-3 句）
  ——然而，关于这个问题，学界目前的理解还存在哪些空白？本文从哪个角度填补？
```

**最容易犯的错误**：第三层写成"本研究具有重要的理论与现实意义"——这句话什么都没说。  
要具体：理论意义是什么？现实意义指向谁？对什么决策有帮助？

---

### 4.2 国内外研究现状

文献综述不是把文章一篇篇列出来，而是**按照研究问题的逻辑分类展示**。

**组织方式（主题分类法）：**
```
2.1 [与你研究相关的维度A] 方面的研究
    ——现有研究发现了什么？用 2-3 篇代表性文献说明。
    ——这些研究的不足是什么？

2.2 [维度B] 方面的研究
    ——同上。

2.3 文献小结
    ——梳理现有研究的整体不足，指向本文的研究定位。
```

**质量要求：**
- 近五年文献占比建议超过 70%
- 每个"不足"必须与你后面的"创新点"一一对应
- 不要只罗列，要评价：它好在哪里、欠缺在哪里

---

### 4.3 研究设计

这是开题报告中技术含量最高的部分，要让评委看到你的研究是**可操作的**。

```
数据部分：
  - 使用什么数据库？时间跨度？样本覆盖？
  - 核心变量怎么测量？有没有现成指标或需要自己构建？

方法部分：
  - 使用什么计量方法？为什么这个方法适合这个问题？
  - 如何处理内生性？（OLS / IV / DiD / RD / PSM 等）

技术路线图：
  用一张流程图展示从数据到结论的完整路径，
  让评委一眼看出你的研究有头有尾。
```

> 如果你的方法部分写得模糊，很可能说明你自己还没想清楚。  
> 开题被挑战最多的往往就是这里。

---

### 4.4 预期贡献与创新点

创新点要具体，不要写"本研究填补了领域空白"这种废话。

**三类常见创新点（任选其一或组合）：**

| 类型 | 示例表述 |
|------|---------|
| 视角创新 | 现有研究多关注企业层面，本文从员工个体层面考察… |
| 方法创新 | 现有文献以 OLS 为主，本文利用 XX 政策作为工具变量，识别因果效应 |
| 数据创新 | 本文使用 XX 数据库，样本比现有研究扩展至 XX 年，覆盖了 XX 新政策周期 |


## 五、开题答辩应对

**评委关心的核心问题无非三个：**
1. 这个问题真的值得研究吗？
2. 你的方法能回答这个问题吗？
3. 你有没有能力完成这个研究？

**高频问题与参考回应：**

| 被问到 | 不要这样说 | 可以这样说 |
|--------|-----------|-----------|
| 内生性怎么处理 | "我会控制更多变量" | "我使用 XX 作为工具变量/准自然实验，理由是…" |
| 数据从哪来 | "我会想办法" | "数据来自 XX 数据库，已获取/已申请" |
| 创新点在哪 | "有一定创新性" | "与 XX 文章相比，我在 A/B/C 维度上做了扩展" |
| 不确定的问题 | 胡说 | "这个问题很好，我目前还没想清楚，我会在后续研究中重点关注" |
