## 一、A 股市场数据

### 行情数据
- 日 / 周 / 月度行情（开盘价、收盘价、最高价、最低价、成交量、换手率）
- 复权因子（前复权 / 后复权 / 不复权）
- 指数行情（沪深 300、中证 500、创业板等主要指数）
- 分钟级行情（1分钟、5分钟、15分钟、30分钟、60分钟）

### 公司基本面
- 财务三表（资产负债表、利润表、现金流量表）——年报 / 季报 / 半年报
- 主要财务指标（ROE、ROA、毛利率、资产负债率等衍生指标）
- 每股指标（EPS、BVPS、每股经营现金流）
- 审计意见类型

### 股权与公司治理
- 股东结构（前十大股东、机构持股比例）
- 控股股东及实际控制人变更
- 股权质押明细
- 高管团队信息（任期、薪酬、持股）
- 独立董事名单与变更

### 资本市场事件
- IPO 上市信息（发行价、超额认购倍数、募资规模）
- 增发 / 配股 / 可转债发行
- 分红送转记录
- 股票停复牌记录
- ST / 退市状态变更


## 二、宏观经济数据

### 国内宏观指标
- GDP（名义 / 实际，季度 / 年度，分省数据）
- CPI / PPI（总指数及分类）
- 工业增加值 / 固定资产投资 / 社会消费品零售总额
- 进出口总额及分类（HS 编码层面）
- PMI（制造业 / 非制造业 / 财新）

### 货币与金融
- M0 / M1 / M2 货币供应量
- 社会融资规模（总量及各分项）
- 银行贷款利率 / 存款利率 / LPR / SHIBOR
- 存款准备金率变化记录
- 人民币汇率（中间价，对主要货币）

### 财政数据
- 全国 / 分省财政收支（一般公共预算）
- 政府性基金收支
- 国债 / 地方债发行规模

### 就业与人口
- 城镇登记失业率 / 调查失业率
- 分省城镇人口 / 户籍人口
- 农民工数量及流向

### 行业与价格
- 工业品出厂价格（PPI 分行业）
- 大宗商品价格（螺纹钢、铜、原油等）
- 房地产价格指数（70 城新建商品住宅）

## 七、数据应用场景

我们提供的数据主要用于**论文实证分析**，涵盖常见计量研究场景：

- **事件研究（Event Study）**：公告日前后的累计超额收益（CAR）
- **面板回归（Panel FE/RE）**：企业 / 城市层面多年平衡面板
- **双重差分（DID）**：政策前后的处理组 / 对照组数据
- **工具变量（IV）**：适合作为外生工具的历史数据、地理数据
- **GARCH / 时序分析**：高频价格序列、宏观月度序列
- **机器学习特征工程**：多维度财务 + 市场特征合并

数据可按需裁剪为**研究所需时间范围 + 样本范围**，并提供清洗后的 CSV / Excel 格式。
