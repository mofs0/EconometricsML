# A05 毕业大论文（学位论文）写作指南

毕业大论文和小论文是完全不同的两种游戏。  
小论文像短跑：创新点突出、方法精准、结论聚焦。  
大论文像马拉松：系统性、完整性、工作量缺一不可。

---


## 一、大论文 vs 小论文的本质差异

| 维度 | 小论文（期刊论文） | 大论文（学位论文） |
|------|------------------|--------------------|
| 评价标准 | 创新性第一 | 系统性 + 工作量 + 创新性 |
| 长度 | 5000-12000 词 | 3-10 万字 |
| 结构 | 相对灵活 | 必须有完整章节（绪论、综述、实证、结论）|
| 读者 | 同行专家 | 答辩委员会（不全是你的细分领域专家）|
| 写法 | 精炼，直接进入主题 | 详细，背景铺垫充分 |

**最重要的启示**：大论文不是把三篇小论文拼在一起——各章之间需要有统一的研究逻辑串联。  
在写大论文之前，先想清楚这几章的**内在逻辑关系**是什么。


## 二、标准章节结构

以经济/管理/社科类为例：

```
封面
中英文摘要
目录
图表目录

第一章  绪论
  1.1  研究背景与问题提出
  1.2  研究意义（理论 + 实践）
  1.3  研究思路与技术路线图
  1.4  研究内容与章节安排
  1.5  本文的创新点

第二章  文献综述与理论基础
  2.1  [方向A] 的研究综述
  2.2  [方向B] 的研究综述
  2.3  理论框架与研究假说

第三章  数据、变量与研究方法
  3.1  数据来源与样本描述
  3.2  变量定义与测量
  3.3  模型设定与识别策略

第四章  实证结果与分析
  4.1  基准回归结果
  4.2  稳健性检验
  4.3  机制检验
  4.4  异质性分析

第五章  进一步分析（第二个研究问题，可选）
  ...

第六章  研究结论与政策建议
  6.1  主要研究结论
  6.2  政策建议
  6.3  研究不足与展望

参考文献
附录
```

> 每所学校的格式要求不同，以学校官方的学位论文撰写规范为准。


## 三、大论文的写作节奏

**最好的策略**：不要等到最后半年才开始写大论文。

**建议的时间安排（三年制硕士）：**

```
第一年
  · 上学期：大量阅读文献，熟悉方法，确定方向
  · 下学期：完成开题报告，开始数据收集和处理
  · 暑假：争取完成第一个研究的初步分析

第二年
  · 上学期：完成第一篇小论文，投出去
  · 下学期：在修改第一篇的同时，开展第二个研究
  · 同期：开始写大论文的第一章（绪论）和第二章（综述）

第三年
  · 上学期：整合各章，形成大论文完整初稿，交导师审
  · 下学期：外审、预答辩、修改、正式答辩
```

> 绪论和文献综述是最容易提前写的部分，放到最后写反而效率最低。  
> 建议第二年就完成这两章的初稿。


## 四、大论文的评分逻辑

答辩委员会打分时看的不是某一部分有多亮眼，而是整体评价。  
了解评分维度，才能有针对性地分配精力。

**核心评分维度：**

**格式规范** —— 这是最基础的要求，不达标会被打回重改。  
按学校模板做，字体、行距、页边距、图表编号都要统一。

**结构完整性** —— 各章之间是否有逻辑递进关系？  
评委会问：你的第三章结论和第四章的分析有什么联系？

**工作量** —— 数据规模、分析深度、稳健性检验数量都是工作量的体现。  
工作量不足是被退改最常见的原因之一。

**文献掌握程度** —— 综述是否全面？是否只写了几篇文章？  
近五年文献的覆盖率是重要指标。

**研究方法的合理性** —— 你为什么选这个方法？内生性怎么处理？  
这是最容易被答辩委员深挖的地方。

**创新性** —— 你做了什么别人没做过的事？  
注意：大论文的"新"侧重系统性的新贡献，不一定要求方法上的颠覆式创新。


## 五、查重与学术规范

**查重的基本规则：**
- 国内学校普遍使用 CNKI 学术不端检测系统（知网查重）
- 各校重复率红线不同，通常在 15%-30% 之间
- 自己已发表的小论文被引用进大论文，会被系统检测为重复——需要提前了解学校的处理政策

**合理降重的方式（不改变意思）：**
- 改变句子结构，用不同的词汇表达同一意思
- 把直接引用改为间接引用（用自己的话概括文献的核心观点）
- 检查是否有大段照搬文献原文的情况，改写为自己的表述

**学术规范的底线：**
- 引用他人数据、图表、代码，必须标注来源
- 不能把他人的研究成果以自己的名义发表
- 论文辅导机构可以提供意见，但不能替你写——一旦查实代写，学位撤销且记入档案

> 学位是你的，风险也是你的。


## 六、答辩准备

**答辩 PPT 结构（以 20-25 分钟为例）：**

```
第一部分（3-4 页）：研究背景与问题
  ——为什么这个问题重要？你的研究出发点是什么？

第二部分（2-3 页）：文献综述与研究定位
  ——前人做了什么？你做了什么不一样的？

第三部分（3-4 页）：数据与研究方法
  ——数据来源 + 核心变量 + 识别策略（重点！）

第四部分（5-6 页）：主要实证结果
  ——基准回归 + 稳健性检验 + 机制/异质性（用图表，不要大段文字）

第五部分（2 页）：结论与贡献
  ——最重要的发现是什么？对学界/实践的启示？
```

**答辩时高频被问的问题：**

```
Q：你的研究如何处理内生性问题？
A：说明工具变量 / 双重差分 / 断点回归 的识别逻辑，以及满足的假设条件

Q：这个结果是否具有外部效度？能推广到其他地区/时期吗？
A：诚实说明样本的适用范围，以及局限性

Q：某个控制变量为什么要加/不加？
A：从理论上解释该变量是否是混淆因素，是否存在过控问题

Q：结论与 XX 文献的发现有出入，怎么解释？
A：从数据、时期、地区差异等维度解释可能的原因
```
