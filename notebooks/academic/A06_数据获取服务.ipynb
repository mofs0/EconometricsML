## 3. 国际数据（免费公开来源）

### 3.1 美国宏观数据（FRED，圣路易斯联储）

**完全免费**，覆盖 80 万+ 指标，是宏观经济研究的标准数据源。

```python
from fredapi import Fred
fred = Fred(api_key='your_free_api_key')  # 官网免费申请

gdp = fred.get_series('GDP')              # 美国实际 GDP
cpi = fred.get_series('CPIAUCSL')         # 美国 CPI
fedfunds = fred.get_series('FEDFUNDS')    # 联邦基金利率
unrate = fred.get_series('UNRATE')        # 失业率
```

**常用指标（部分）：**
VIX 恐慌指数、10年期国债利率、美元指数、原油价格、黄金价格、各国汇率……

## 6. 数据获取示范（免费可运行）

以下代码可直接运行，展示我们的数据获取能力：


````python
# 示范：用 AKShare 获取免费宏观数据（无需 token）
# pip install akshare

try:
    import akshare as ak
    import pandas as pd
    
    # 中国 GDP 年度数据
    df_gdp = ak.macro_china_gdp_yearly()
    print("✅ 中国GDP年度数据（最近5年）：")
    print(df_gdp.tail())
    print()
    
    # A股上证指数日线行情
    df_sh = ak.stock_zh_index_daily(symbol="sh000001")
    df_sh['date'] = pd.to_datetime(df_sh['date'])
    df_sh = df_sh[df_sh['date'] >= '2020-01-01']
    print("✅ 上证指数日线数据（2020年至今，前5行）：")
    print(df_sh.head())

except ImportError:
    print("请先安装 akshare：pip install akshare")
except Exception as e:
    print(f"数据获取异常（网络原因）：{e}")
````

````python
# 示范：用 yfinance 获取国际股票数据（免费）
# pip install yfinance

try:
    import yfinance as yf
    import pandas as pd
    
    # 标普 500 近 3 年
    sp500 = yf.download('^GSPC', start='2022-01-01', progress=False)
    print("✅ 标普500 日线数据（最近5行）：")
    print(sp500[['Open','High','Low','Close','Volume']].tail())
    
    # 港股腾讯
    tencent = yf.download('0700.HK', start='2022-01-01', progress=False)
    print(f"\n✅ 港股腾讯 数据行数：{len(tencent)}")

except ImportError:
    print("请先安装：pip install yfinance")
except Exception as e:
    print(f"获取异常：{e}")
````
