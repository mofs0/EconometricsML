# A01 文献检索与高效阅读

**定位**：适合刚入学的硕博生，帮助建立文献工作的系统性方法。

文献工作贯穿整个研究周期，但多数人的做法是低效的：漫无目的地收藏、读了忘、找不到思路。  
本篇的目标是让你用更少的时间，从文献中提取更多有用的东西。

---


## 一、先想清楚：你为什么要读文献

读文献不是一件有固定答案的事情，目的不同，方法就应该不同。  
入学初期和写论文中期，读文献的姿势应该完全不一样。

**四种常见的读文献场景：**

| 场景 | 目标 | 读法 |
|------|------|------|
| 刚进入一个新领域 | 建立整体认知 | 找 3-5 篇综述，快速浏览，建立地图 |
| 寻找研究方向 | 发现空白或矛盾 | 精读近 3 年高引文章，重点看 Limitation |
| 遇到具体技术问题 | 找解决方案 | 定向搜索，只看 Methods 部分 |
| 写作阶段找素材 | 支撑论点 | 提取核心结论，不需要全文读 |

> **一句话原则**：带着问题去读，没有问题就先不要读。


## 二、去哪里找文献

### 国内文献平台

- **知网（CNKI）**：中文期刊最全，学校账号免费，是中文文献的首选
- **万方 / 维普**：知网的补充，特定期刊有独家收录

### 国际文献平台

- **Web of Science（WOS）**：SCI/SSCI 的权威检索入口，学校账号必开
- **Google Scholar**：覆盖面最广，但不区分期刊质量，适合"撒网"
- **Semantic Scholar**：有 AI 摘要和引用图谱，适合快速判断一篇文章是否值得读
- **Connected Papers**：输入一篇文章，自动生成关联文献网络图，找同方向文献的神器

### 下载渠道

| 情况 | 解决方案 |
|------|---------|
| 学校有订阅 | 直接在期刊官网下载 |
| 学校没有 | Sci-Hub（粘贴 DOI 即可） |
| 想找书籍/报告 | Library Genesis（libgen.rs）|
| 想联系作者 | Google 找邮箱直接发邮件，回复率通常 >70% |

> 🔍 **高效找文献的节奏**：Scholar 撒网 → WOS 确认质量 → Sci-Hub 下载 → Zotero 收藏


## 三、关键词怎么搭

多数人用一个词搜，结果要么太多要么找不到想要的。

**三层搜索法：**

```
第一层（宽）：大词 + "review" 或 "survey"
  → 快速找综述，建立领域地图
  
第二层（中）：核心方法词 + 应用场景词
  → 例：instrumental variable + wage inequality
  
第三层（精）：精确词组，用引号锁定
  → 例："regression discontinuity" + "education policy" + China
```

**找到一篇好文章后的滚雪球技巧：**
1. 往前看：它的参考文献列表（经典文章都在这里）
2. 往后看：WOS 里的"Times Cited"，谁引用了它（最新进展在这里）
3. 横向看：同一作者的其他论文（同方向高手的研究路径）


## 四、读文献的核心框架

读完一篇文章，你应该能回答这三个问题：

```
① 它研究了什么问题？（Why this matters）
② 它发现了什么？   （What the findings are）
③ 它怎么做到的？   （How it was done）
```

基于这三个问题，推导出第四个：

```
④ 它哪里没做好？我能在哪里往前走一步？
```

这个第四问，就是你的创新点的来源。

**读文献的时间分配（每篇精读文章）：**
```
标题 + 摘要（5 分钟）  → 决定要不要精读
Introduction（15 分钟） → 理解研究动机和贡献
Conclusion（10 分钟）  → 确认主要发现
Methods（20 分钟）     → 理解识别策略/技术路线
Results（15 分钟）     → 看关键表格和图
```

> 绝大多数文章只需要读摘要和结论。精读是奢侈品，用在最值得的文章上。


## 五、找创新点：文献不足的常见类型

读了很多文章，但不知道从哪里入手写自己的研究？  
把下面这张清单当作检查表，逐条问一遍：

| 类型 | 说明 | 你能做什么 |
|------|------|-----------|
| **地理空白** | 现有研究集中在欧美，中国/发展中国家缺数据 | 用中国数据复现或对比 |
| **时间空白** | 数据只到某年，新政策/新事件未被研究 | 延伸时间窗口 |
| **维度缺失** | 只研究了 A 对 B 的影响，没考虑 C 的调节 | 加入调节变量/异质性分析 |
| **机制黑箱** | 知道 X 影响 Y，但不知道为什么 | 做机制检验（中介分析）|
| **方法陈旧** | 用 OLS，但存在明显内生性问题 | 用 IV/DiD/RD 重新识别 |
| **数据粗糙** | 用年度数据，但问题本质是短期的 | 用月度或日度数据 |

> 你的创新不需要颠覆一个领域，只需要让这个领域向前移动一小步。


## 六、文献管理：用 Zotero 建立知识库

**为什么推荐 Zotero（免费开源）：**
- 浏览器插件一键保存，自动抓取元数据
- Word / LaTeX 双插件，写论文时直接插入引用
- 标签 + 文件夹分类，几百篇文献也不乱

**基本工作流：**
```
找到文章 → Zotero Connector 一键保存 
→ 阅读时在 Zotero 内做高亮/笔记
→ 写作时直接插入引用（格式自动生成）
→ 参考文献列表自动排版
```

**给每篇精读文章写一份简短读书笔记（模板）：**
```
论文：[作者年份标题]
研究问题：
主要方法：
核心发现：
可借鉴之处：
不足/可改进之处：
```
