## 一、变量赋值

```python
# 用 = 号把右边的值存进左边的变量
nian_ling = 22
name = "Zhang San"
GPA = 3.85

print(nian_ling)
print(name)
print(f"GPA: {GPA}")  # f-string：花括号里放变量名自动替换
```

## 二、四种基本数据类型

| 类型 | 英文 | 例子 | 经济学用途 |
|------|------|------|------------|
| 整数 | `int` | `2024` | 年份、样本量 |
| 浮点数 | `float` | `3.14` | 价格、比率、系数 |
| 字符串 | `str` | `"GDP"` | 变量名、标签 |
| 布尔值 | `bool` | `True` | 条件判断、虚拟变量 |

```python
# type() 查看数据类型
print(type(2024))       # int
print(type(0.05))       # float
print(type("renjun GDP"))  # str
print(type(True))       # bool

# 数字运算
print(10 + 3)   # 13
print(10 / 3)   # 3.333...
print(10 // 3)  # 3  整除
print(10 % 3)   # 1  取余
print(10 ** 2)  # 100 幂次
```

## 三、列表 list：一排数据

```python
# 创建列表
year = [2019, 2020, 2021, 2022, 2023]
gdp_growth = [6.0, 2.3, 8.1, 3.0, 5.2]

# 索引从 0 开始
print(year[0])    # 2019
print(year[-1])   # 2023（最后一个）
print(year[1:3])  # [2020, 2021] 切片

# 常用操作
print(len(year))          # 长度 5
print(sum(gdp_growth))    # 求和
print(max(gdp_growth))    # 最大值
year.append(2024)         # 追加
print(year)
```

## 四、字典 dict：有名字的数据格

用花括号 `{}` 表示，每项数据有一个键名，类比 Excel 表格的列名。

```python
# 用字典存储一家企业的财务数据
company = {
    "name": "CATL",
    "year": 2023,
    "revenue": 4009.17,   # 亿元
    "net_profit": 441.21,
    "total_assets": 6383.44,
    "debt_ratio": 0.627
}

# 访问
print(company["name"])
print(company["net_profit"])

# 计算衍生指标
ROE = company["net_profit"] / (company["total_assets"] * (1 - company["debt_ratio"]))
print(f"ROE = {ROE:.2%}")

# 添加新字段
company["ROE"] = round(ROE, 4)
print(company)
```

## 五、练习题

创建一个字典存储任意一家上市公司的财务数据（至少含：名称、年份、营业收入、净利润），然后计算并打印净利润率。

```python
# 参考答案
firm = {
    "name": "Kweichow Moutai",
    "year": 2023,
    "revenue": 1505.60,
    "net_profit": 747.34
}
margin = firm["net_profit"] / firm["revenue"]
print(f"{firm['name']} {firm['year']} net margin: {margin:.2%}")
```
