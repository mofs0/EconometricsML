## 一、定义函数

```python
# def 函数名(参数):
#     函数体
#     return 返回值

def net_profit_margin(net_profit, revenue):
    """Calculate net profit margin (net_profit / revenue)."""
    if revenue == 0:
        return None
    return net_profit / revenue

result = net_profit_margin(441.21, 4009.17)
print(f"Net margin: {result:.2%}")

# 参数默认值
def print_summary(name, profit, revenue, year=2023):
    m = net_profit_margin(profit, revenue)
    print(f"[{year}] {name} | margin: {m:.2%}")

print_summary("CATL", 441.21, 4009.17)
print_summary("Moutai", 747.34, 1505.60, year=2023)
```

## 二、读懂别人的函数（docstring）

```python
def sharpe_ratio(returns, rf=0.03, ann=252):
    """
    Calculate Sharpe ratio.

    Parameters
    ----------
    returns : list  daily return series
    rf      : float annualized risk-free rate, default 3%
    ann     : int   annualization factor, 252 for daily data

    Returns
    -------
    float  Sharpe ratio
    """
    import statistics
    mean_r = statistics.mean(returns)
    std_r  = statistics.stdev(returns)
    return (mean_r * ann - rf) / (std_r * (ann ** 0.5))

help(sharpe_ratio)
```

## 四、调用 empirlab 跑 OLS 模型

```python
# 确保在项目根目录下（或已 pip install -e .）
from empirlab.traditional.ols import OLS, mincer_data

# 生成模拟数据（Mincer 工资方程）
df = mincer_data(n=300)
print("Data shape:", df.shape)
print(df.head())
```

```python
# 拟合 OLS
X = df[["educ", "exper", "exper2"]]
y = df["ln_wage"]

model = OLS(robust=True).fit(X, y)
print(model.summary_table().round(4))
print(f"\nR2 = {model.r_squared:.4f}  |  N = {model.nobs}")
print("\nResult: educ coef ~0.10 means 1 more year of education -> ~10% higher wage")
```

## 五、练习题

写函数 `leverage(assets, equity)` 返回财务杠杆倍数，对 5 家企业批量计算：

```python
def leverage(assets, equity):
    """Financial leverage = total assets / total equity."""
    return assets / equity

firms = [
    ("Moutai",  2723.89, 1962.45),
    ("CATL",    6383.44, 2375.05),
    ("CMB",  110174.0, 10583.0),
    ("Gree",    3309.88,  811.60),
    ("BYD",     7025.63, 2017.21),
]

for name, assets, equity in firms:
    lev = leverage(assets, equity)
    print(f"{name:10s}  leverage: {lev:.2f}x")
```
