## 一、Jupyter Notebook 是什么？

Jupyter Notebook 就像一个**可以运行代码的 Word 文档**，页面由一个个单元格 Cell 组成，分两种：

| 类型 | 作用 | 识别方法 |
|------|------|----------|
| **Markdown 单元格** | 显示说明文字 | 单击后左侧无 `[ ]` |
| **Code 单元格** | 写代码并运行 | 左侧有 `[ ]` 或数字 |

**运行一个单元格：** 点击它，然后按 `Shift + Enter`

````python
# 试运行这里：点击此单元格，按 Shift + Enter
print("你好，Python！")
print("运行成功，恭喜完成第一步！")
````

## 二、最常用快捷键

| 快捷键 | 作用 |
|--------|------|
| `Shift + Enter` | 运行并跳到下一格 |
| `Ctrl + Enter` | 运行并停在原地 |
| `A` | 上方插入新单元格（先按 Esc）|
| `B` | 下方插入新单元格 |
| `D D` | 删除当前单元格 |
| `Ctrl + S` | 保存 |

## 三、怎么看报错信息

报错很正常，关键是**看最后一行**。格式：`错误类型: 原因`

常见错误：
- `NameError: name 'x' is not defined` → 变量 x 没有赋值
- `SyntaxError` → 语法写错了，比如少了冒号或括号
- `ModuleNotFoundError` → 包没安装，运行 `pip install 包名`
- `ZeroDivisionError` → 除以零了

````python
# 这行会报错，运行看看长什么样
try:
    result = 1 / 0
except ZeroDivisionError as e:
    print(f'报错类型：ZeroDivisionError')
    print(f'报错原因：{e}')
    print('处理方法：检查分母是否为零')
````
