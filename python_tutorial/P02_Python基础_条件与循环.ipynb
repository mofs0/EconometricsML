## 一、if / elif / else

```python
profit = -500

if profit > 0:
    print("Profitable")
elif profit == 0:
    print("Break even")
else:
    print("Loss")

# 比较运算符：>  <  >=  <=  ==  !=
# 逻辑运算符：and  or  not
```

```python
# 多条件判断
debt_ratio = 0.75
net_margin = -0.02

if debt_ratio > 0.70 and net_margin < 0:
    print("High risk: high leverage + loss")
elif debt_ratio > 0.70:
    print("Warning: high leverage")
elif net_margin < 0:
    print("Warning: loss-making")
else:
    print("Financially healthy")
```

## 二、for 循环

```python
# 遍历列表
industries = ["Manufacturing", "Finance", "IT", "Healthcare", "Consumer"]
for ind in industries:
    print(f"Industry: {ind}")

print("---")

# range() 固定次数
for i in range(5):
    print(f"Year {2019 + i}")
```

```python
# 批量计算资产负债率并筛选高风险企业
firms = [
    {"name": "Firm A", "assets": 1000, "liabilities": 800},
    {"name": "Firm B", "assets": 500,  "liabilities": 150},
    {"name": "Firm C", "assets": 2000, "liabilities": 1600},
    {"name": "Firm D", "assets": 800,  "liabilities": 300},
]

high_risk = []
for firm in firms:
    dr = firm["liabilities"] / firm["assets"]
    firm["debt_ratio"] = dr
    flag = "High risk" if dr > 0.70 else "OK"
    print(f"{firm['name']:8s}  debt ratio: {dr:.1%}  [{flag}]")
    if dr > 0.70:
        high_risk.append(firm["name"])

print(f"\nHigh risk firms: {high_risk}")
```

## 三、列表推导式

```python
returns = [0.05, -0.02, 0.08, -0.01, 0.12]

# 普通写法
pos = []
for r in returns:
    if r > 0:
        pos.append(r)
print("Normal:", pos)

# 列表推导式（等价，更简洁）
pos = [r for r in returns if r > 0]
print("Comprehension:", pos)

# 带计算的推导式
labels = [f"{r:.1%}" for r in returns]
print("Labels:", labels)
```

## 四、练习题

给定以下 ROE 数据，用循环找出 ROE > 15% 的企业：

```python
roe_data = {"Moutai":34.2, "Ping An":10.1, "CATL":22.7, "CMB":16.8,
            "Gree":28.3, "Alibaba":8.5, "BYD":18.9, "PetroChina":9.3}

# 参考答案
print("ROE > 15%:")
for name, roe in roe_data.items():
    if roe > 15:
        print(f"  {name}: {roe}%")

high_roe = [n for n, r in roe_data.items() if r > 15]
print(f"\nTotal: {len(high_roe)} firms: {high_roe}")
```
