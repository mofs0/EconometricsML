## 1. 基本设置


```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# 支持中文显示（Windows）
plt.rcParams['font.family'] = ['Microsoft YaHei', 'SimHei', 'DejaVu Sans']
plt.rcParams['axes.unicode_minus'] = False  # 负号正常显示
plt.rcParams['figure.dpi'] = 100

print('Matplotlib 准备好了！')
```

---
## 2. 折线图：GDP 增速时间序列


```python
# 中国 GDP 增速（模拟数据，近似真实趋势）
years  = list(range(2000, 2024))
gdp_gr = [8.5, 8.3, 9.1, 10.0, 10.1, 11.3, 12.7, 14.2,
           9.6,  9.2, 10.6, 9.5,  7.9,  7.8,  7.3,  6.9,
           6.7,  6.9,  6.6,  6.1,  2.3,  8.1,  3.0,  5.2]

fig, ax = plt.subplots(figsize=(10, 4))
ax.plot(years, gdp_gr, marker='o', linewidth=2, color='steelblue', label='GDP增速')
ax.axhline(y=6, color='red', linestyle='--', alpha=0.5, label='6%参考线')
ax.fill_between(years, gdp_gr, 6, where=[g >= 6 for g in gdp_gr],
                alpha=0.1, color='green')

ax.set_title('中国GDP年增速（2000-2023）', fontsize=14, pad=12)
ax.set_xlabel('年份')
ax.set_ylabel('增速（%）')
ax.legend()
ax.grid(axis='y', linestyle=':', alpha=0.6)
plt.tight_layout()
plt.show()
```

## 3. 散点图：工资与教育年限


```python
np.random.seed(42)
edu_yrs = np.random.choice([9, 12, 15, 16, 19], 200,
                            p=[0.15, 0.30, 0.10, 0.35, 0.10])
# Mincer工资方程：ln(wage) = a + b*edu + 噪声
log_wage = 6.5 + 0.08 * edu_yrs + np.random.normal(0, 0.4, 200)
wage = np.exp(log_wage)

fig, axes = plt.subplots(1, 2, figsize=(11, 4))

# 左图：原始散点
axes[0].scatter(edu_yrs, wage, alpha=0.5, edgecolors='none', color='coral')
axes[0].set_title('教育年限 vs 月工资')
axes[0].set_xlabel('教育年限（年）')
axes[0].set_ylabel('月工资（元）')

# 右图：对数工资 + 回归线
axes[1].scatter(edu_yrs, log_wage, alpha=0.5, edgecolors='none', color='teal')
b = np.polyfit(edu_yrs, log_wage, 1)
x_line = np.linspace(8, 20, 100)
axes[1].plot(x_line, np.polyval(b, x_line), 'r-', linewidth=2,
             label=f'斜率={b[0]:.3f}')
axes[1].set_title('教育年限 vs 对数工资（Mincer方程）')
axes[1].set_xlabel('教育年限（年）')
axes[1].set_ylabel('ln(月工资)')
axes[1].legend()

plt.tight_layout()
plt.show()
print(f'教育年限每增加1年，工资增加约 {b[0]*100:.1f}%（半对数估计）')
```

## 4. 直方图：收益率分布


```python
np.random.seed(0)
returns = np.random.normal(0.001, 0.02, 1000)  # 模拟日收益率

fig, ax = plt.subplots(figsize=(8, 4))
ax.hist(returns * 100, bins=40, color='steelblue', edgecolor='white',
        alpha=0.8, density=True)

# 叠加正态曲线
from scipy import stats
x = np.linspace(returns.min()*100, returns.max()*100, 200)
ax.plot(x, stats.norm.pdf(x, returns.mean()*100, returns.std()*100),
        'r-', linewidth=2, label='正态分布')
ax.axvline(0, color='black', linestyle='--', alpha=0.5)

ax.set_title('日收益率分布（1000个交易日）')
ax.set_xlabel('日收益率（%）')
ax.set_ylabel('密度')
ax.legend()
plt.tight_layout()
plt.show()
```

## 5. 箱线图：分行业 ROE 比较


```python
np.random.seed(7)
industries = ['制造业', '金融业', '零售业', '房地产', '信息技术']
means = [0.08, 0.12, 0.07, 0.10, 0.15]
data_by_ind = [np.random.normal(m, 0.06, 100) for m in means]

fig, ax = plt.subplots(figsize=(9, 4))
bp = ax.boxplot(data_by_ind, labels=industries, patch_artist=True,
                medianprops=dict(color='red', linewidth=2))

colors = ['#4C72B0','#DD8452','#55A868','#C44E52','#8172B2']
for patch, color in zip(bp['boxes'], colors):
    patch.set_facecolor(color)
    patch.set_alpha(0.7)

ax.set_title('各行业 ROE 分布对比（箱线图）')
ax.set_ylabel('ROE')
ax.axhline(0, color='gray', linestyle=':', alpha=0.5)
ax.grid(axis='y', linestyle=':', alpha=0.5)
plt.tight_layout()
plt.show()
```

## 6. 图表美化小技巧
| 设置项 | 代码 |
|--------|------|
| 标题 | `ax.set_title('...', fontsize=14)` |
| 坐标轴标签 | `ax.set_xlabel('...')` |
| 图例 | `ax.legend(loc='upper left')` |
| 网格线 | `ax.grid(linestyle=':', alpha=0.5)` |
| 保存图片 | `plt.savefig('fig.png', dpi=300, bbox_inches='tight')` |
| 图形大小 | `plt.subplots(figsize=(宽, 高))` |
